```python
# Define parameters
max_depth = 3
strategy = 'fos'
max_speed = 8
speed_importance = 1
session_name = f'{max_depth}_{strategy}_{max_speed}_{speed_importance}'
save = 1

# Create directory for trips
if save:
    os.makedirs(f'../trips/{session_name}', exist_ok=True)

# List of track numbers
#tracks = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10']
tracks = ['10']

# Execute the main loop
for track_no in tracks:
    # Start timing
    start_time = time.time()
    track1 = Track(f'../tracks/track_{track_no}.t')
    racecar1 = RaceCar(track1, max_depth=max_depth,strategy=strategy,max_speed=max_speed,speed_importance=speed_importance)
    racecar1.complete_moves()
    race1 = Race(racecar1, track1,draw_distances=1)

    # End timing
    end_time = time.time()

    # Calculate the elapsed time
    elapsed_time = end_time - start_time
    
    performance = len(racecar1.pos_hist) - 1
    print(f'{track_no}: {performance}\nTotal execution time: {elapsed_time:.2f} seconds\n\n')
    
    if save:
        race1.save_trip(f'../trips/{session_name}/trip{track_no}.rl')

        f = open('../trips/' + session_name + '/details.txt', "w")
        f.write(f'{track_no}: {performance}\nTotal execution time: {elapsed_time:.2f} seconds\n\n')
        f.close()
```
